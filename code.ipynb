```python
import pandas as pd 
import matplotlib.pyplot as pt 
import seaborn as sea 
import plotly.express as ple 
```

```python
df=pd.read_csv('country_wise_latest.csv')
```

```python
df.head()
```

### Region VS Confirmed Cases

```python
df.groupby('WHO Region')['Confirmed'].sum().sort_values(ascending= False).plot(kind='bar')
pt.title('Region vs Confirmed Cases')
pt.xlabel('Region')
pt.ylabel('Confirmed Cases')
pt.xticks(rotation=45)
pt.grid(axis='y')
pt.show()
```

### The Top 10 Country with height Active Cases

```python
pt.figure(figsize=(10,6))
df.groupby('Country/Region')['Active'].sum().sort_values(ascending= False).head(10).plot(kind='bar')
pt.title('Top 10 Country with height Active Cases')
pt.xlabel('Country')
pt.ylabel('Active Cases')
pt.xticks(rotation=45)
pt.grid(True)
pt.show()
```

### Image show  for Confirmed & Deaths & Recovered & Active cases

```python
pt.figure(figsize=(10,6))
cl=df[['Confirmed','Deaths','Recovered','Active']].corr()
pt.imshow(cl,cmap='Blues_r')
pt.xticks(range(len(cl.columns)),cl.columns,rotation=45)
pt.yticks(range(len(cl.columns)),cl.columns)
pt.colorbar()
pt.show()
```

### Cases Fatality VS Cases Recovered

```python
a=df['Fatality_Rate']=df['Deaths']/df['Confirmed']*100
b=df['Recovered_Rate']=df['Recovered']/df['Confirmed']*100
```

```python
pt.figure(figsize=(10,6))
sea.scatterplot(x=a,y=b)
pt.title('Cases Fatality VS Cases Recovered')
pt.xlabel('Fatality')
pt.ylabel('Recovered')
pt.grid(True)
```

### Top 10 Country with height Active Cases

```python
pt.figure(figsize=(10,6))
df.groupby('Country/Region')['New cases'].sum().sort_values(ascending= False).head(10).plot(kind='pie', autopct='%1.2f%%',)
pt.title('Top 10 Country with height Active Cases')
pt.xlabel('Country')
pt.ylabel('Active Cases')
pt.xticks(rotation=45)
pt.grid(True)
pt.show()
```

### Cases Deaths VS Cases Recovered

```python
pt.figure(figsize=(10,6))
sea.scatterplot(df,x='Deaths',y='Recovered')
pt.title('Cases Deaths VS Cases Recovered')
pt.xlabel('Deaths')
pt.ylabel('Recovered')
pt.grid(True)
```

### Top 10 Country with Maximun Covid Reporting

```python
pt.figure(figsize=(10,6))
df1=df.groupby('Country/Region')['Deaths'].sum().sort_values(ascending= False).head(10)
pt.title('Top 10 Country with Maximun Covid Reporting')
pie_Chart=pt.pie(df1,labels=df1.index,autopct='%.2f%%')
pt.show()
```

### Top 4 Country  With Percentage Deaths

```python
pt.figure(figsize=(10,6))
df2=df.groupby('Country/Region')['Deaths / 100 Cases'].sum().sort_values(ascending= False).head(4)
colors=['red','green','gold','violet'] 
pt.xlabel('Deaths / 100 Cases')
pt.title('Top 4 Country  With Percentage Deaths')
pt.barh(df2.index,df2,color=colors)
pt.show()
```
